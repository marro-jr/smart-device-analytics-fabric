### **Paso 1 - Leer la tabla "category" usando "Spark.sql"**

```python
category_df = spark.sql("select * from lh_silver.dbo.category")
```

### **Paso 2 - Eliminar columna "environment" del DataFrame**

```python
from pyspark.sql.functions import col
```

```python
category_final_df = category_df.drop(col("environment"))
```

### **Paso 3 - Escribir datos en el "lh_gold"**

```python
category_final_df.write.format("delta").mode("overwrite").option("overwriteSchema", "true").saveAsTable("dim_category")
```

```python
%%sql
SELECT * FROM dim_category limit 10;
```
