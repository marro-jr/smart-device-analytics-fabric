```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType
```

```python
connectivity_schema = StructType(fields= [
    StructField("ConnectivityID",IntegerType(), False),
    StructField("USB",StringType(),True),
    StructField("USBConnector",StringType(),True),
    StructField("Bluetooth",StringType(),True),
    StructField("NFC",StringType(),True),
    StructField("SIMCardSlot",StringType(),True)
])
```

```python
connectivity_df = spark.read.schema(connectivity_schema).option("multiLine",True).json(f"{bronze_folder_path}/{p_file_date}/connectivity.json")
```

```python
from pyspark.sql.functions import lit
```

```python
connectivity_final_df = (
    add_ingestion_date(connectivity_df, p_file_date)
    .withColumnRenamed("ConnectivityID", "connectivity_id")
    .withColumnRenamed("USB", "usb")
    .withColumnRenamed("USBConnector", "usb_connector")
    .withColumnRenamed("Bluetooth", "bluetooth")
    .withColumnRenamed("NFC", "nfc")
    .withColumnRenamed("SIMCardSlot", "sim_card_slot")
    .withColumn("envrionment", lit(p_environment))
)
```

```python
display(connectivity_final_df.limit(5))
```

```python
connectivity_final_df.write.format("delta").mode("overwrite").option("overwriteSchema", "true").saveAsTable("connectivity")
```
